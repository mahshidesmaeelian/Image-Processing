```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Input/Carrot.jpg")
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
R , G , B = cv2.split(img)
plt.imshow(G , cmap= "gray")
```

```python
new_R = G
new_G = G
new_B = B
```

```python
img_result = cv2.merge((new_B , new_G , new_R))
cv2.imwrite("Output/Green_carrot.jpg" , img_result)
img_result = cv2.cvtColor(img_result , cv2.COLOR_BGR2RGB)
plt.imshow(img_result)
```
