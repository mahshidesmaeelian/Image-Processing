```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Input/rubix.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#converting yellow parts to blue color
img[np.where((img == [255, 255, 0]).all(axis = 2))] = [0 , 0 , 255]
#converting cyan parts to red color
img[np.where((img == [0, 255, 255]).all(axis = 2))] = [255 , 0 , 0]
#converting magenta parts to green color
img[np.where((img == [255, 0, 255]).all(axis = 2))] = [0 , 255 , 0]
```

```python
plt.imshow(img)
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
cv2.imwrite("Output/Rubix_result.jpg" , img)
```
