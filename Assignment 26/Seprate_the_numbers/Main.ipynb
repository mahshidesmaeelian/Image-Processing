```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Seprate_the_numbers/mnist.png")
img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
img = cv2.equalizeHist(img)
plt.imshow(img , cmap="gray")
```

```python
width , height = img.shape
print(width , height)
```

```python
cell_width = width// 50
cell_height = height// 100
print(cell_width , cell_height)
```

```python
# PATH = "output"
for i in range(10):
    try:
        os.mkdir(f"Seprate_the_numbers/Num{i}")
    except FileExistsError:
        print("this folder already exist")
```

```python
count = 0
f = 0

for i in range(0 , width , cell_width):
    for j in range(0 , height , cell_height):

        small_img = img[i:i+cell_width , j:j+cell_height]

        if small_img.shape == (cell_width , cell_height):
            cv2.imwrite(f"Seprate_the_numbers/Num{f}/{count}.jpg" , small_img)
            count+=1

            if count >= 500:
                f += 1
                count = 0
```
